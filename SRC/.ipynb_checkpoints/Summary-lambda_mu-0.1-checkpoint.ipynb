```python
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data_sgd = {}
data_baseline = {}
for year in range(1991, 2020):
    y_str = str(year)
    with open('./data/loss_lam0p01_{}.pkl'.format(y_str), 'rb') as f:
        try:
            while True:
                if year not in data_sgd:
                    data_sgd[year] = []
                team, loss = pickle.load(f)
                data_sgd[year].append((team, loss.item()))
        except EOFError:
            pass

    
    with open('./data/loss-baseline_lam0p01_{}.pkl'.format(y_str), 'rb') as f:
        try:
            while True:
                if year not in data_baseline:
                    data_baseline[year] = []
                team, loss = pickle.load(f)
                data_baseline[year].append((team, loss))
        except EOFError:
            pass
```

```python
def get_loss(data, name):
    loss = pd.DataFrame()
    for year in range(1991, 2020):
        if len(data[year]) != 0:
            if loss.empty:
                loss = pd.DataFrame.from_dict(data[year])
                loss.columns = ['Team', 'Loss_'+name]
                loss['Year'] = year
            else:
                tmp = pd.DataFrame.from_dict(data[year])
                tmp.columns = ['Team', 'Loss_'+name]
                tmp['Year'] = year        
                loss = pd.concat([loss, tmp])
    return loss

loss_sgd = get_loss(data_sgd, 'SGD')
loss_baseline = get_loss(data_baseline, 'baseline')
```

```python
win_rate = pd.DataFrame()
cols = ['Team', 'W/L%']

for year in range(1991, 2020):
    y_str = str(year)

    path1 = '../NBA/data/win_rate/win_rate_W_{}.csv'.format(y_str)
    path2 = '../NBA/data/win_rate/win_rate_E_{}.csv'.format(y_str)

    
    win_rate1 = pd.read_csv(path1)[cols]
    win_rate2 = pd.read_csv(path2)[cols]
    tmp = pd.concat([win_rate1, win_rate2])
    tmp["Year"] = year
    if win_rate.empty:
        win_rate = tmp
    else:
        win_rate = pd.concat([win_rate, tmp])
win_rate = win_rate.merge(loss_sgd, on = ['Team', 'Year'], how = 'right')    
win_rate = win_rate.merge(loss_baseline, on = ['Team', 'Year'], how = 'left')
win_rate
```

```python
win_rate['I'] = ((win_rate['Loss_SGD'] - win_rate['Loss_baseline'])/10000)
win_rate['I_norm'] = ((win_rate['Loss_SGD'] - win_rate['Loss_baseline'])/np.abs(win_rate['Loss_baseline']))
```

Define SGD loss functions, $F^*$.  
Define ground truth loss function, $F$.  

$F^*\left(\mu^*,\{T\}|\nu,\{K\}\right)=\left( \sum_l<-\text{ln}(K^{l}),T^l>-\frac{1}{\lambda}H(T^l)-\frac{1}{\lambda_{\mu}}H(\mu^*) \right)$,  
where $H(T^l)=-\sum_{i,j}T^l_{i,j}(\text{ln}(T^l_{i,j})-1)$ and $H(\mu^*)=-\sum_{i}\mu_i(\text{ln}(\mu^*)-1)$ are the entropic regularizations. $\lambda$ and $\lambda_{\mu}$ are hyperparaters set to 1. $\mu^*$ is the resulted row marginal from performing SGD on $\{K^{(y,t)}\}$. $T^l=SK^L(\mu^*, \nu, K^l)$ is the $L$-step Sinkhorn approximation.  

$F\left(\mu^g,\tilde{T^g},\nu^g,\{K\}\right)=\left( \sum_l<-\text{ln}(K^{l}),\tilde{T^g}>-\frac{1}{\lambda}H(\tilde{T^g})-\frac{1}{\lambda_{\mu}}H(\mu) \right)$,  
where $\mu^g,\nu^g$, and, $\tilde{T^g}$ are the ground truth marginals and coupling.

Define index $I = F^*/N-F/N$. $N$ is the sample size of $\{K\}$.

Assume the PIOT inferred $\{K\}$ reflect the players' ability, EOT and SGD find a solution of the marginal, $\mu^*$, which minimizes the total loss $F^*$. Such $\mu^*$ is the marginal when the teammates cooperate in a best way. The figure shows the correlation between the win rate and $I$. We find that $F^* < F$ (Since $I<0$.) for all teams from 1991 to 2019 as shown in the figure.
