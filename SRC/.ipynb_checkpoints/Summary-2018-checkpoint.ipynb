### Define collaborative index. (Tentative)

[1]   
$I(y,t)=\sum_i (max(r_i^{(y,t)})-1)/N^{(y,t)}$,  
where $r_i^{(y,t)}$ is row $i$ of the observed coupling and $N^{(y,t)}$ is the number of players of team $t$ in year $y$.  

**Pros:** Identify the max element of each row and quantify the concentration of the matrix.   
**Cons:** Might not be ideal if players hare good at multiple actions.  

[2] KL divergence  
$I(T^{(y,t)},\{K^{(y,t)}\}) = \sum_{k} \sum_{i,j} \text{KL}(T^{(y,t)} || K_{i,j}^{(y,t,k)}) / N^{(y,t)}$,  
where $K_i^{(y,t)}$ is the inferred PIOT Gibbs cost and $T^{(y,t)}$ is the normalized observed coupling.

**Pros:** Take into account PIOT simulaitons. Measure the difference between $K_{i,j}^{(y,t,k)} = T^{(y,t)}$.   
**Cons:** Do not measure the concertration of T.  

[3] Cross entropy  
$I(T^{(y,t)},\{K^{(y,t)}\}) = - \sum_{k} \sum_{i,j} T^{(y,t)} * \text{ln} (K_{i,j}^{(y,t,k)}) / N^{(y,t)}$,

**Pros:** Has minimum when $K_{i,j}^{(y,t,k)} = T^{(y,t)}$, which is the entropy of $T^{(y,t)}$. Measure the concertration (or uncertainty) of $T^{(y,t)}$.   
**Cons:** While number of players is different for each team, how to compare across team? 


For [3], lower value of I indicates either the coupling has high concentration or the average of Gibbs cost is close to the coupling, or both. Could compute the entropy of $T^{(y,t)}$ for further investigation. Lowest possible entropy is 0.

$F(K, \alpha, \beta)=<-ln(K), T> - H(T)$

$\alpha^*, T^* = argmin_{\alpha, T} \text{F}(T, \alpha | K^{(y-1)}, \beta)$

$F^* < F$.(?) $\frac{F^*(\alpha^{y,*}, T^{y,*})-F(\alpha^y, T^y)}{F}$ more negative means worse cooperation and better win-rate/points/ranking.

Why does $F^*$ represent best cooperation?

```python
import pickle
import numpy as np

data_sgd = []
with open('loss-lambda1.pkl', 'rb') as f:
    try:
        while True:
            data_sgd.append(pickle.load(f))
    except EOFError:
        pass
    
data_baseline = []
with open('loss-baseline.pkl', 'rb') as f:
    try:
        while True:
            data_baseline.append(pickle.load(f))
    except EOFError:
        pass
```

```python
dict_sgd = {}
dict_baseline = {}

for team, loss in data_sgd:
    dict_sgd[team] = loss
for team, loss in data_baseline:
    dict_baseline[team] = loss
```

```python
I = {}
for key in dict_sgd:
    I[key] = np.abs((dict_sgd[key]-dict_baseline[key])/dict_baseline[key]).numpy()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(figsize =(8,6))
data = [list(I.values())]
ax = fig.add_axes([0, 0, 1, 1])
x = np.ones(len(data[0]))
bp = ax.scatter(x, data)
plt.show()
```

```python
path1 = '../NBA/data/win_rate/win_rate_W_2019.csv'
path2 = '../NBA/data/win_rate/win_rate_E_2019.csv'


win_rate1 = pd.read_csv(path1)
win_rate2 = pd.read_csv(path2)
win_rate = pd.concat([win_rate1, win_rate2]).reset_index()
loss = pd.DataFrame.from_dict(I, orient = 'index')
loss.index.names = ['Team']
loss = loss.rename(columns={0: "I"})
win_rate = win_rate.join(loss, on = 'Team')
win_rate
```

```python
win_rate.plot.scatter(x='I', y='W/L%')
```
