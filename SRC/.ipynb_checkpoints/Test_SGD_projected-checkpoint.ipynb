```python
def row_iter(K, b, row_sum):
    return row_sum / torch.matmul(K, torch.transpose(b, 0, -1))

def col_iter(K, a, col_sum):
    return col_sum / torch.matmul(torch.transpose(K, 0, 1), torch.transpose(a, 0, -1))

def sinkhorn_torch_base(K,
                        row_sum,
                        col_sum,
                        num_iter
                        ):
    '''
    Sinkhorn scaling base
    Parameters
    ----------
    mat     : muted torch 2-tensor of shape(n,m)
    row_sum : immuted torch 1-tensor of size n
    col_sum : immuted torch 1-tensor of size m
    epsilon : tolerance of 1-norm on column-sums with rows normalized
    max_iter: maximal iteration steps (multiples of 10)
    Return
    ------
    Sinkhorn scaled matrix 
    '''
    
    a = torch.ones(len(row_sum))
    b = torch.ones(len(col_sum))

    while num_iter:
        a = row_iter(K, b, row_sum)
        b = col_iter(K, a, col_sum)

        num_iter -= 1

    return torch.matmul(torch.matmul(torch.diag(a), K.clone()), torch.diag(b))

def sinkhorn_torch_fix_steps(mat,
                   row_sum=None,
                   col_sum=None,
                   num_iter = 20,

                   # row_check=False, # Not activate for efficiency concern
                  ):
    '''
    Sinkhorn scaling base
    Parameters
    ----------
    mat     : muted torch 2-tensor of shape(n,m)
    row_sum : immuted torch 1-tensor of size n
    col_sum : immuted torch 1-tensor of size m
    epsilon : tolerance of 1-norm on column-sums with rows normalized
    max_iter: maximal iteration steps (multiples of 10)
    row_confident: whether row sums are all nonzeros (to bypass the check for nan-problem)
    col_confident: whether col sums are all nonzeros (to bypass the check for nan-problem)
    data: stores row and col sums during sinkhorn iterations
    Return
    ------
    Sinkhorn scaled matrix (in place, can skip capturing it)
    '''
    n, m = mat.shape
    row_sum = row_sum if row_sum is not None else torch.ones(n, dtype=torch.float64,
                                                             device=mat.device)
    col_sum = col_sum if col_sum is not None else torch.ones(m, dtype=torch.float64,
                                                             device=mat.device)

    # if row_check:
    #     mat[row_sum != 0, col_sum != 0] = sinkhorn_torch_base(mat[row_sum != 0, col_sum != 0],
    #                                                           row_sum[row_sum != 0],
    #                                                           col_sum[col_sum != 0],
    #                                                           epsilon, max_iter)
    # mat[:, col_sum == 0.] = 0.
    # mat[row_sum == 0., :] = 0.
    mat2 = mat.clone()
    mat2[:, col_sum != 0.] = sinkhorn_torch_base(mat2[:, col_sum != 0.],
                                                row_sum,
                                                col_sum[col_sum != 0],
                                                num_iter)
    mat2[:, col_sum == 0.] = 0.

    return mat2

def sinkhorn_loss(K, r, c, L, nr, nc):
    T = sinkhorn_torch_base(K.clone(),
                   row_sum=r,
                   col_sum=c,
                   num_iter = L)
    return ( -torch.log(K)*T ).sum() #+ 0.001*(r*(torch.log(r)-torch.ones(len(r)))).sum() 


def init(nr, nc, mean, std, num_matrices):
    matrices = []

    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    #torch.manual_seed(42)
    r = torch.ones(nr, requires_grad=True, dtype=torch.float, device=device)
    #r = torch.normal(size = (nr,), mean = 1.0, std = 0.01, requires_grad=True, dtype=torch.float, device=device)
    c = torch.ones(nc, requires_grad=False, dtype=torch.float, device=device)

    for _ in range(num_matrices):
        # Dirichlet ---
        Dir = Dirichlet(0.5*torch.ones((nr, nc)))
        new_mat = Dir.sample()
        
        # Gaussian ---
        #new_mat = torch.normal(mean = mean, std = std, size = (nr, nc))
        #new_mat = torch.normal(mean = mean*torch.ones(nr), std = std)
        #while new_mat.min() < 0.:
        #    new_mat = torch.normal(mean = mean, std = std, size = (nr, nc))     
        
        # Diagonal ---
        #new_mat = torch.diag(new_mat)
        #adds = torch.ones((nr, nc))*1e-15
        #new_mat += adds

        matrices.append(new_mat)

    print(r)
    print(c)
    print(matrices[0])
    print('\n')
    
    return r, c, matrices
```

## Diagonal matrix, uniform initial

```python
nr, nc = 3, 3
mean, std = 0.3, 1e-1
L = 100
lr = 0.01
num_matrices = 100
num_epoch = 50

# r, c, matrices = init(nr, nc, mean, std, num_matrices)

matrices = []

device = 'cuda' if torch.cuda.is_available() else 'cpu'
#torch.manual_seed(42)
r = torch.tensor([0.5, 1.0, 1.5], requires_grad=True, dtype=torch.float, device=device)
#r = torch.normal(size = (nr,), mean = 1.0, std = 0.01, requires_grad=True, dtype=torch.float, device=device)
c = torch.ones(nc, requires_grad=False, dtype=torch.float, device=device)

for _ in range(num_matrices):   

    # Diagonal ---
    diag = torch.ones(nr)
    new_mat = torch.diag(diag)
    adds = torch.ones((nr, nc))*1e-15
    new_mat += adds

    matrices.append(new_mat)

print(r)
print(c)
print(matrices[0])
print('\n')

optimizer = torch.optim.RMSprop([r], lr=lr)

data_loss = []
data_r = []
data_c = []

for epoch in range(num_epoch):
    running_loss = 0
    for i in range(len(matrices)):
        loss = sinkhorn_loss(matrices[i].clone(), r, c, L, nr, nc)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        with torch.no_grad():
            for param in [r, c]:
                param[:] = param.clamp(1e-15, 10)
                param[:] = param / param.sum() * nr
        
        running_loss += loss.item()
        
    data_loss.append(running_loss)
    data_r.append(r.cpu().detach().clone().numpy())
    data_c.append(c.cpu().detach().clone().numpy())
        
    print('Epoch {}'.format(epoch))
    print('Loss: {}'.format(running_loss))
    print('r: {}'.format(r))
    print(r.sum().detach().numpy())
    print('c: {}'.format(c))
    print('\n')
    
print(data_r[-1])
print(data_r[-1].sum())

fig, ax = plt.subplots(1, 2, figsize = (10, 5))

ax[0].plot(data_loss)
ax[0].set_title('Loss')
ax[1].plot(data_r)
ax[1].set_title('Marginal')
fig.tight_layout()
```
